```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
wine = pd.read_csv("D:\\Data Science\\0 - ASSIGNMENTS\\Assignment8 -PCA\\wine.csv")
wine.head()
```

```python
wine.info()
```

```python
# Considering only numerical data 
wine.data = wine.iloc[:,1:]
wine.data.head()
```

```python
# Converting into numpy array
WINE = wine.data.values
WINE
```

```python
# Normalizing the numerical data 
wine_norm = scale(WINE)
```

```python
wine_norm
```

```python
# Applying PCA Fit Transform to dataset
pca = PCA()
pca_values = pca.fit_transform(wine_norm)
```

```python
pca_values
```

```python
pca = PCA(n_components = 13)
pca_values = pca.fit_transform(wine_norm)
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pca.components_
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
```

```python
 pca_values[:,0:1]
```

```python
x = pca_values[:,0:1]
y = pca_values[:,1:2]
plt.scatter(x,y)
```

```python
df=pd.concat([wine['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
df
```

#### Hierachial clustering with PCA

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
df1=df.iloc[:,1:]
df1
```

```python
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(df1,'complete'))
```

```python
# Create Clusters
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hc
```

```python
y_hc = hc.fit_predict(df1)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters.head()
```

```python
Clusters.value_counts()
```

```python
wine['Type'].unique()
```

##### The original data has 3 cluster types and the Hierarchial Clustering after performing PCA also has 3 Clusters

#### K-means clustering in PCA

```python
from sklearn.cluster import KMeans
```

```python
df1
```

```python
wcss=[]
for i in range(1,10):
    kMeans=KMeans(n_clusters=i,random_state=0)
    kMeans.fit(df1)
    wcss.append(kMeans.inertia_)
    
plt.plot(range(1, 10), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(3, random_state=42)
clusters_new.fit(df1)
```

```python
clusters_new.labels_
```

```python
# Assign clusters to the data set
wine1=wine.copy()
wine1['clusters']=clusters_new.labels_
wine1
```

```python
wine1['clusters'].value_counts()
```

##### The original data has 3 cluster types and the K Means Clustering after performing PCA also has 3 Clusters
