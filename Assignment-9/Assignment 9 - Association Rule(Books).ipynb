#### Prepare rules for the all the data sets 
#### 1) Try different values of support and confidence. Observe the change in number of rules for different support,confidence values,2) Change the minimum length in apriori algorithm,3) Visulize the obtained rules using different plots 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
book=pd.read_csv('D:\\Data Science\\0 - ASSIGNMENTS\\Assignment9 - Association Rules\\book.csv')
book
```

```python
#Apriori Algorithm
```

```python
#1 - Association rule with 10% support and 50% confidence
```

```python
frequent1=apriori(book,min_support=0.1,use_colnames=True)
frequent1
```

```python
rules1=association_rules(frequent1,metric='lift',min_threshold=0.5)
rules1
```

```python
rules1.sort_values('lift',ascending=False)
```

```python
rules1[rules1.lift>1]
```

```python
plt.scatter(rules1['support'],rules1['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```
